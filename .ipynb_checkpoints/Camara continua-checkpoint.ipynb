```python
import cv2
import time


cam = cv2.VideoCapture(0)  


directorio = "/home/carlitosduty/Documents/codes/NeuralNetworks/ReconocimientoFacial/data/img/myimg/"


try:
    contador = 1
    while True:
        
        ret, imagen = cam.read()

     
        nombre_archivo = f'imagen_{contador+78}.jpg'
        cv2.imwrite(directorio + nombre_archivo, imagen)

        
        print(f'Imagen {contador} tomada')

        
        contador += 1

        
        time.sleep(1)

except KeyboardInterrupt:
    cam.release()
    cv2.destroyAllWindows()
    time.sleep(1)
```
